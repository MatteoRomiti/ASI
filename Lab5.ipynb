#### 2.1.1 Scoring 
<div class="alert alert-info">
Use the score described in the lectures (only on the training data) to get a set of features (corresponding to a subset of the original ones). You should make sure you extract this subset from the test data too!
</div>

<div class="alert alert-success">
Let's create some functions to compute the scores of each feature and then order them by score.
</div>

```python
def get_feature_score(X_feat, t):
    """
    X_feat: array of N scalars corresponding to one input feature
    t: labels 
    """
    X_0 = np.array([X_feat[i] for i in range(len(X_feat)) if t[i]==0])
    X_1 = np.array([X_feat[i] for i in range(len(X_feat)) if t[i]==1])
    s = np.abs(X_1.mean() - X_0.mean())/(X_0.var() - X_1.var())
    return s

def order_features_by_score(X):
    scores = []
    for feat in range(X.shape[1]):
        scores.append((feat, get_feature_score(X[:,feat], t))) # [feature_number, feature_score]
    scores.sort(key=lambda x : x[1], reverse=True)
    return np.array([x[0] for x in scores]), np.array([x[1] for x in scores])
```

```python
ordered_features, scores = order_features_by_score(X)
print("Ordered features")
for i in range(10):
    print("Feature:", ordered_features[i], "\tScore:", scores[i])
```

```python
plt.scatter(X_test[:,ordered_features[0]], X_test[:,ordered_features[1]], c=t_test)
# plt.scatter(X[:,ordered_features[0]], X[:,ordered_features[1]], c=t)
plt.title("Black=Leukemia - White=Healthy")
plt.show()
```

<div class="alert alert-success">
When using the best 2 features computed on the training set, we don't get very good results when projecting because the two classes are quite mixed in the center of the figure. Let's try with 3 features.
</div>

```python
fig = plt.figure(1, figsize=(12, 8))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, 1, 1], elev=35, azim=75)
plt.cla()
ax.scatter(X_test[:,ordered_features[0]], X_test[:,ordered_features[1]], \
           X_test[:,ordered_features[3]], c=t_test, s=40)
plt.title("Black=Leukemia - White=Healthy")
plt.show()
```

<div class="alert alert-success">
Even with 3 features, the problem remains: the two classes are mixed in the center of the 3D space.

#### 2.1.3 Principal components
<div class="alert alert-info">
You can find a PCA implementation in sklearn. 
</div>

# 2D Visualization

```python
pca = decomposition.PCA(n_components=2) # specify the number of components
pca.fit(X) # fit the model with X
X_pca = pca.transform(X) #apply dimensionality reduction to X
print("Output shape:",X_pca.shape)
plt.scatter(X_pca[:,0], X_pca[:,1], c=t)
plt.title("Black=Leukemia - White=Healthy")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

<div class="alert alert-success">
This result shows the power and usefulness of PCA. The two classes have been split quite well, but this is on the training set. Let's try on the test set.
</div>

```python
pca = decomposition.PCA(n_components=2) # specify the number of components
pca.fit(X) # fit the model with X
X_pca = pca.transform(X_test) #apply dimensionality reduction to X_test
plt.scatter(X_pca[:,0], X_pca[:,1], c=t_test)
plt.title("Black=Leukemia - White=Healthy")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

<div class="alert alert-success">
It's not a perfect result, but much better than the feature selection by scoring.
</div>

# 3D Visualization

```python
pca = decomposition.PCA(n_components=3) # specify the number of components
pca.fit(X) # fit the model with X
X_pca = pca.transform(X_test) #apply dimensionality reduction to X_test

fig = plt.figure(1, figsize=(15, 10))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, 1, 1], elev=25, azim=45)
plt.cla()
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=t_test, s=100)
plt.title("Black=Leukemia - White=Healthy")
plt.show()
```

<div class="alert alert-success">
Similar comment as above: the two classes have been split quite well, but some (2) leukemia points are very close to the healthy class.
