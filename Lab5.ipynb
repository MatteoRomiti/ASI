```python
from sklearn import decomposition
```

# 2D Visualization

```python
print("Input shape:",X.shape)
pca = decomposition.PCA(n_components=2) # specify the number of components
pca.fit(X) # fit the model with X
X_pca = pca.transform(X) #apply dimensionality reduction to X
print("Output shape:",X_pca.shape)
plt.scatter(X_pca[:,0], X_pca[:,1], c=t)
plt.title("Dataset")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

# 3D Visualization

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets

print("Input shape:",X.shape)
pca = decomposition.PCA(n_components=3) # specify the number of components
pca.fit(X) # fit the model with X
X_pca = pca.transform(X) #apply dimensionality reduction to X
print("Output shape:",X_pca.shape)

fig = plt.figure(1, figsize=(15, 10))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, 1, 1], elev=45, azim=45)
plt.cla()
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=t, s=40, cmap=plt.cm.jet)

plt.show()
```
